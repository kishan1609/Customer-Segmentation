```python
# Feature Engineering
```

```python
# Dt_Customer that indicates the date a customer joined the database is not parsed as DateTime

df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
```

```python
print("The newest customer's enrollment date in the records: ", max(df["Dt_Customer"]))
print("The oldest customer's enrollment date in the records: ", min(df["Dt_Customer"]))
```

```python
# Extract the "Age" of a customer by the "Year_Birth" indicating the birth year of the respective person.

df["Age"] = 2015 - df["Year_Birth"]
```

```python
# Create another feature "Spent" indicating the total amount spent by the customer in various categories over the span of two years.

df['Spent'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
```

```python
# Create another feature "Living_With" out of "Marital_Status" to extract the living situation of couples.

df['Living_With'] = df['Marital_Status'].replace({'Married':'Partner', 'Together':'Partner', 'Absurd':'Alone', 'Widow':'Alone', 'YOLO':'Alone', 'Divorsed':'Alone', 'Single':'Alone'})
```

```python
# Create a feature "Children" to indicate total children in a household that is, kids and teenagers.

df['Children'] = df['Kidhome'] + df['Teenhome']
```

```python
# To get further clarity of household, Creating feature indicating "Family_Size"

df['Children'] = df['Children'].astype(int)
df['Family_Size'] = df['Living_With'].replace({'Alone': 1, 'Partner': 2,  'Divorced': 1, 'Married': 2}).astype(int) + df['Children']
```

```python
# Create a feature "Is_Parent" to indicate parenthood status

import numpy as np
df['Is_Parent'] = np.where(df.Children > 0, 1, 0)
```

```python
# Segmenting education levels in three groups

df['Education'] = df['Education'].replace({'Basic':'Undergraduate', '2n Cycle':'Graduate', 'Master':'Postgraduate', 'Phd':'Postgraduate'})
```

```python
# Dropping some of the redundant features

to_drop = ['Marital_Status', 'Dt_Customer', 'Z_CostContact', 'Z_Revenue', 'Year_Birth', 'ID']
df = df.drop(to_drop, axis=1)
```

```python
numerical = ['Income', 'Recency', 'Age', 'Spent']

def detect_outliers(d):
  for i in d:
    Q3, Q1 = np.percentile(df[i], [75 ,25])
    IQR = Q3 - Q1

    ul = Q3+1.5*IQR
    ll = Q1-1.5*IQR

    outliers = df[i][(df[i] > ul) | (df[i] < ll)]
    print(f'*** {i} outlier points***', '\n', outliers, '\n')
```

```python
# creating a subset of dataframe by dropping the features on deals accepted and promotions
cols_del = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response']
df = df.drop(cols_del, axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(df)
df = pd.DataFrame(X, columns = df.columns)
df.head(3)
```

```python
# Model Building
```

```python
# 1. PCA
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95, random_state=42) # I retained 95% variance to reduce noise while preserving cluster structure.
x_pca = pca.fit_transform(X)
```

```python
# Remove near-zero variance features

from sklearn.feature_selection import VarianceThreshold

vt = VarianceThreshold(threshold=0.01)
x_var = vt.fit_transform(x_pca)
```

```python
# 2. KMeans
```

```python
# Elbow Method
```

```python
from sklearn.cluster import KMeans

inertia = []

K = range(2, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(x_var)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,5))
plt.plot(K, inertia, marker='o', )
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
# K = 4
# Reasoning (how to explain in interview / project):
# There is a steep decrease in inertia from K = 2 to K = 4
# After K = 4, the curve starts to flatten
```

```python
# Final Kmeans model
from sklearn.metrics import silhouette_score

k_means = KMeans(n_clusters=4, random_state=42, n_init=50)
kmeans_pred = kmeans.fit_predict(x_var)

kmeans_sil = silhouette_score(x_var, kmeans_pred)
print("Kmeans Silhouette Score", round(kmeans_sil, 3))
```

```python
# 3. DBSCAN
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=1.2, min_samples=5)
dbscan_pred = dbscan.fit_predict(x_pca)

# remove noise points (-1) before silhouette 
mask = dbscan_pred != -1

if len(set(dbscan_pred[mask])) > 1:
    dbscan_sil = silhouette_score(x_pca[mask], dbscan_pred[mask])
else:
    dbscan_sil = -1

print("DBSCAN Silhouette Score:", round(dbscan_sil, 3))
```

```python
# 4. Hierarchical Clustering(Agglomerative)
```

```python
from sklearn.cluster import AgglomerativeClustering

hier = AgglomerativeClustering(n_clusters=4, linkage="ward")
hier_pred = hier.fit_predict(x_var)

hier_sil = silhouette_score(x_var, hier_pred)
print("Hierarchical Silhouette Score:", round(hier_sil, 3))
```

```python
print("MODEL COMPARISON")
print("----------------")
print(f"KMeans       : {kmeans_sil:.3f}")
print(f"DBSCAN       : {dbscan_sil:.3f}")
print(f"Hierarchical : {hier_sil:.3f}")
```

```python
# Hyperparameter Tuning(Grid search)
```

```python
# For KMeans

best_kmeans_score = -1
best_kmeans_params = None

for k in range(2, 11):  # try 2 to 10 clusters
    kmeans = KMeans(
        n_clusters=k,
        init='k-means++',
        n_init=20,
        random_state=42
    )
    
    kmeans_labels = kmeans.fit_predict(x_pca)
    score = silhouette_score(x_pca, kmeans_labels)
    
    if score > best_kmeans_score:
        best_kmeans_score = score
        best_kmeans_params = k

print("Best KMeans Silhouette Score:", round(best_kmeans_score, 3))
print("Best number of clusters (k):", best_kmeans_params)
```

```python
# For DBSCAN

best_score = -1
best_params = None

for eps in [0.5, 0.7, 0.9, 1.1, 1.3]:
    for min_samples in [4, 5, 8, 10]:
        dbscan_labels = DBSCAN(eps=eps, min_samples=min_samples).fit_predict(x_pca)
        mask = dbscan_labels != -1
        
        if len(set(dbscan_labels[mask])) > 1:
            score = silhouette_score(x_pca[mask], dbscan_labels[mask])
            if score > best_score:
                best_score = score
                best_params = (eps, min_samples)

print("Best DBSCAN Score:", round(best_score, 3))
print("Best eps, min_samples:", best_params)
```

```python
# For Hierarchical

from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

best_hier_score = -1
best_hier_params = None

for k in range(2, 11):
    for linkage in ['ward', 'complete', 'average']:
        
        # ward requires euclidean distance
        if linkage == 'ward':
            model = AgglomerativeClustering(
                n_clusters=k,
                linkage=linkage
            )
        else:
            model = AgglomerativeClustering(
                n_clusters=k,
                linkage=linkage,
                metric='euclidean'
            )
        
        hier_labels = model.fit_predict(x_pca)
        score = silhouette_score(x_pca, hier_labels)
        
        if score > best_hier_score:
            best_hier_score = score
            best_hier_params = (k, linkage)

print("Best Hierarchical Silhouette Score:", round(best_hier_score, 3))
print("Best (n_clusters, linkage):", best_hier_params)
```

```python
print("FINAL OPTIMIZED COMPARISON")
print("--------------------------")
print(f"KMeans       : {best_kmeans_score:.3f}")
print(f"DBSCAN       : {best_score:.3f}")
print(f"Hierarchical : {best_hier_score:.3f}")
```

```python
# This checks whether DBSCAN is stable and meaningful.
```

```python
# DBSCAN with best params
final_dbscan = DBSCAN(eps=best_params[0], min_samples=best_params[1])
dbscan_labels = final_dbscan.fit_predict(x_pca)

# Cluster distribution
cluster_counts = pd.Series(dbscan_labels).value_counts().sort_index()
print("DBSCAN Cluster Distribution:")
print(cluster_counts)

# Noise percentage
noise_pct = np.mean(dbscan_labels == -1) * 100
print(f"\nNoise percentage: {noise_pct:.2f}%")
```

```python
# KMeans distribution
kmeans_counts = pd.Series(kmeans_labels).value_counts()
print("KMeans cluster sizes:")
print(kmeans_counts)

# Hierarchical distribution
hier_counts = pd.Series(hier_labels).value_counts()
print("\nHierarchical cluster sizes:")
print(hier_counts)
```

```python
# FINAL MODEL: KMEANS

# DBSCAN âŒ (99% noise)
# Hierarchical âŒ (99% in one cluster)
# KMeans âœ… (balanced, meaningful clusters)

final_KMeans = KMeans(n_clusters=4, n_init=50, random_state=42)
KMeans_labels = final_KMeans.fit_predict(x_pca)
KMeans_sil = silhouette_score(x_pca, KMeans_labels)
KMeans_sil
```
